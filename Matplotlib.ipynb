2.<b> Anatomy of a Matplotlib Plot</b>
<p style="font-family:Helvica; font-size:21px; ">
Matplotlib has a flexible structure that allows you to control each part of a plot individually. Let’s look at the core components:
<p style="font-family:Helvica; font-size:21px; ">
<b>Figure: </b>The overall container for one or more plots (axes). Think of it as the "canvas" where everything is drawn.</br>
<b>Axes: </b>A specific plot inside a figure where data is plotted. It is essentially a "subplot." Each figure can contain multiple axes.</br>
<b>Axis: </b>Refers to the individual x-axis and y-axis within an axes. These axes help define the plot area, tick marks, and grid.</br>
<b>Title: </b>The title of the entire plot or subplot.</br>
<b>Labels: </b>Labels for the x-axis and y-axis.</br>
<b>Legend: </b>Provides context for different elements within the plot, such as colors representing specific data categories.
</br></br>Let's see these elements in a sample plot:</p>

```python
import matplotlib.pyplot as plt

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 6))  # Creates a figure with a single set of axes

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Plotting the data
ax.plot(x, y, label="Sample Data", color="blue", marker="o")  # Adding label for legend

# Setting title and labels
ax.set_title("Sample Plot")  # Title of the plot
ax.set_xlabel("X-axis Label")  # X-axis label
ax.set_ylabel("Y-axis Label")  # Y-axis label

# Adding legend
ax.legend(loc="upper left")  # Displays the legend in the upper-left corner

# Display the plot
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
Explanation of Each Line:
</br>
<p style="font-family:Helvica; font-size:21px; ">
<li style="font-family:Helvica; font-size:21px; "><b>import matplotlib.pyplot as plt: </b> Imports Matplotlib’s plotting module.
<li style="font-family:Helvica; font-size:21px; "><b>fig, ax = plt.subplots(figsize=(8, 6)): </b> Creates a figure and a set of axes with an 8x6 inches canvas.
<li style="font-family:Helvica; font-size:21px; "><b>x and y:</b> Define data for the x and y axes.
<li style="font-family:Helvica; font-size:21px; "><b>ax.plot(...): </b>Plots the data with a label and customization.
<li style="font-family:Helvica; font-size:21px; "><b>ax.set_title, ax.set_xlabel, ax.set_ylabel:</b>Add titles and labels.
<li style="font-family:Helvica; font-size:21px; "><b>ax.legend(...): </b>Displays the legend to identify the line.


<p style="font-family:Helvica; font-size:24px; ">
3.<b> Coordinate Systems in Matplotlib </b>
<p style="font-family:Helvica; font-size:21px; ">
Matplotlib offers three primary coordinate systems, each useful for different scenarios:
<li style="font-family:Helvica; font-size:21px; "><b>Data Coordinates: </b></br> The default. The coordinates are defined by the data being plotted. This is where the x and y data values are located.
<li style="font-family:Helvica; font-size:21px; "><b>Axes Coordinates: </b> </br>Defines positions relative to the axes, where (0,0) is the bottom-left of the plot area and (1,1) is the top-right. Useful for positioning elements like text labels relative to the plot dimensions.
<li style="font-family:Helvica; font-size:21px; "><b>Figure Coordinates: </b></br> Defines positions relative to the entire figure, where (0,0) is the bottom-left of the figure and (1,1) is the top-right. This is helpful for global figure annotations.
</br></br>Example Using Different Coordinate Systems for Annotations

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Plotting data
ax.plot(x, y, color="blue", marker="o")

# Data Coordinate Annotation (at a specific data point)
ax.annotate("Data Point", xy=(3, 25), xytext=(3.5, 27),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Axes Coordinate Annotation (top left corner of the plot)
ax.text(0.05, 0.95, "Axes Coord Text", transform=ax.transAxes,
        verticalalignment='top', color="green")

# Figure Coordinate Annotation (bottom left of the figure)
fig.text(0.05, 0.05, "Figure Coord Text", color="red")

plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
Explanation of Each Annotation:

<li style="font-family:Helvica; font-size:21px; "><b>Data Coordinates: </b> ax.annotate("Data Point", xy=(3, 25), ...) positions text based on the actual data point (3, 25).
<li style="font-family:Helvica; font-size:21px; "><b>Axes Coordinates: </b> ax.text(..., transform=ax.transAxes) places text relative to the axes (top-left corner of the plot area).
<li style="font-family:Helvica; font-size:21px; "><b>Figure Coordinates: </b> fig.text(...) places text relative to the entire figure area.
</br></br>Each coordinate system allows us to tailor the placement of labels and annotations, enhancing readability and context.

1. <b>Basic Plot Types </b>
<p style="font-family:Helvica; font-size:21px; ">
Matplotlib offers a variety of plots for different kinds of data. Here’s how to create some of the most common ones.</br></br>
<b>a. Line Plot: </b></br>
A line plot is ideal for showing trends over time or ordered data points.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the line plot
plt.plot(x, y, marker="o", linestyle="-", color="blue", label="Trend Line")
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="upper left")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>b. Scatter Plot: </b></br>
A scatter plot shows individual data points and is useful for examining the relationship between two variables.

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Scatter plot
plt.scatter(x, y, color="red", label="Data Points")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>c. Bar Plot: </b></br>
A bar plot is suitable for comparing quantities across categories.

```python
# Data
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 9]

# Bar plot
plt.bar(categories, values, color="purple")
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>d. Histogram </b></br>
A histogram is useful for showing the distribution of a single variable by dividing it into bins.

```python
import numpy as np

# Data
data = np.random.normal(0, 1, 1000)  # Normally distributed data

# Histogram
plt.hist(data, bins=30, color="green", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

3. <b>Additional Plot Types for Exploratory Analysis</b>
<p style="font-family:Helvica; font-size:21px; ">
<b>a. Boxplot</b></br>
A boxplot is useful for visualizing the distribution of data, showing the median, quartiles, and potential outliers.

```python
# Data for boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Plotting boxplot
plt.boxplot(data, vert=True, patch_artist=True, tick_labels=["Group 1", "Group 2", "Group 3"])
plt.title("Boxplot Example")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
In a boxplot:</br>
<p style="font-family:Helvica; font-size:21px; ">
<li style="font-family:Helvica; font-size:21px; ">The box shows the interquartile range (IQR), which contains 50% of the data.
<li style="font-family:Helvica; font-size:21px; ">The line inside the box represents the median.
<li style="font-family:Helvica; font-size:21px; ">Whiskers extend to the smallest and largest values within 1.5 times the IQR. Points outside this range are potential outliers.</br></br>
<b>b. Error Bars</b></br>
Error bars are useful for showing the variability in data, often used to represent standard deviation or confidence intervals.

```python
# Data for error bars
x = np.arange(1, 6)
y = np.array([5, 7, 8, 5, 6])
error = np.array([0.5, 0.8, 0.6, 0.4, 0.7])

# Plot with error bars
plt.errorbar(x, y, yerr=error, fmt="o", color="black", ecolor="red", capsize=5, label="Error Bars")
plt.title("Plot with Error Bars")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
In this example:</br>
<li style="font-family:Helvica; font-size:21px; ">yerr = error specifies the error values.
<li style="font-family:Helvica; font-size:21px; ">fmt = "o" indicates circular markers.
<li style="font-family:Helvica; font-size:21px; ">ecolor = "red" sets the color of error bars.
<li style="font-family:Helvica; font-size:21px; ">capsize = 5 sets the size of the caps on each end of the error bars.</br></br>

1.<b> Creating Multiple Plots in a Single Figure using plt.subplot()</b>
<p style="font-family:Helvica; font-size:21px; ">
The plt.subplot() function is a simple way to create multiple plots in a grid layout within a single figure. It takes three arguments:
<li style="font-family:Helvica; font-size:21px; ">The number of rows.
<li style="font-family:Helvica; font-size:21px; ">The number of columns.
<li style="font-family:Helvica; font-size:21px; ">The index of the subplot you want to select (starting from 1).</br>
Here's a basic example with a 2x2 grid:

```python
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))  # Set the overall figure size

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, index 1
plt.plot([1, 2, 3], [1, 4, 9], 'r-')
plt.title("Plot 1")

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, index 2
plt.bar([1, 2, 3], [1, 4, 9], color="blue")
plt.title("Plot 2")

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, index 3
plt.scatter([1, 2, 3], [1, 4, 9], color="green")
plt.title("Plot 3")

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, index 4
plt.hist([1, 1, 2, 3, 3, 3, 4, 4, 4, 4], bins=5, color="purple")
plt.title("Plot 4")

plt.tight_layout()  # Adjusts spacing between subplots
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>plt.subplot(2, 2, index):</b> Creates subplots in a 2x2 grid.
<li style="font-family:Helvica; font-size:21px; "><b>plt.tight_layout():</b> Adjusts the spacing between subplots to prevent overlapping.</br></br>
<p style="font-family:Helvica; font-size:24px; ">
2.<b> Using plt.subplots() for Flexible Subplot Layouts</b>
<p style="font-family:Helvica; font-size:21px; ">
For more control over individual subplots, plt.subplots() is often preferred as it provides an array of axes objects you can work with. This makes it easier to apply settings and customizations to each subplot individually.

```python
# Create a 2x2 grid of subplots with plt.subplots()
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns, set figure size

# Accessing each subplot through the axes array
axes[0, 0].plot([1, 2, 3], [1, 4, 9], 'r-')
axes[0, 0].set_title("Plot 1")

axes[0, 1].bar([1, 2, 3], [1, 4, 9], color="blue")
axes[0, 1].set_title("Plot 2")

axes[1, 0].scatter([1, 2, 3], [1, 4, 9], color="green")
axes[1, 0].set_title("Plot 3")

axes[1, 1].hist([1, 1, 2, 3, 3, 3, 4, 4, 4, 4], bins=5, color="purple")
axes[1, 1].set_title("Plot 4")

plt.tight_layout()  # Adjusts spacing between subplots
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>fig, axes = plt.subplots(2, 2, figsize=(10, 8)):</b></br> Creates a 2x2 grid of subplots and allows each subplot to be referenced through the axes array.
<li style="font-family:Helvica; font-size:21px; "><b>axes[i, j].set_title():</b></br> Sets titles and other properties individually for each subplot.</br></br>
<p style="font-family:Helvica; font-size:24px; ">
3.<b> Controlling Layout with plt.tight_layout() and figsize</b>
<li style="font-family:Helvica; font-size:21px; ">figsize: Sets the overall dimensions of the figure, making all subplots larger or smaller.
<li style="font-family:Helvica; font-size:21px; ">plt.tight_layout(): Automatically adjusts padding between subplots to avoid overlap.</br></br>
<b>Example:</b>

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, ax in enumerate(axes.flat):  # Iterating through each subplot
    ax.plot([0, 1, 2], [i, i + 1, i + 2])  # Simple plot with incremental values
    ax.set_title(f"Plot {i+1}")
plt.tight_layout()  # Optimizes spacing between subplots
plt.show()
```

4. <b>Advanced Layout Control with GridSpec</b>
<p style="font-family:Helvica; font-size:21px; ">
GridSpec provides a more flexible way to create complex subplot layouts. With it, you can control the row and column spans for each subplot.

```python
import matplotlib.gridspec as gridspec

# Create a figure and a GridSpec layout
fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(3, 3, figure=fig)  # Define a 3x3 grid

# Define subplots with custom spans
ax1 = fig.add_subplot(gs[0, :])  # Row 0, all columns (1x3)
ax2 = fig.add_subplot(gs[1, :-1])  # Row 1, columns 0 and 1 (1x2)
ax3 = fig.add_subplot(gs[1:, -1])  # Row 1 to end, last column (2x1)
ax4 = fig.add_subplot(gs[-1, 0])  # Last row, first column (1x1)
ax5 = fig.add_subplot(gs[-1, -2])  # Last row, middle column (1x1)

# Plot data
ax1.plot([1, 2, 3], [1, 4, 9], 'r-')
ax1.set_title("Plot 1")

ax2.bar([1, 2, 3], [1, 4, 9], color="blue")
ax2.set_title("Plot 2")

ax3.scatter([1, 2, 3], [1, 4, 9], color="green")
ax3.set_title("Plot 3")

ax4.hist([1, 1, 2, 3, 3, 3, 4, 4, 4, 4], bins=5, color="purple")
ax4.set_title("Plot 4")

ax5.plot([1, 2, 3], [9, 6, 3], color="orange", marker="o")
ax5.set_title("Plot 5")

plt.tight_layout()
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>gridspec.GridSpec(3, 3):</b> Creates a 3x3 grid layout for the figure.
<li style="font-family:Helvica; font-size:21px; "><b>gs[0, :]:</b> Occupies all columns in the first row for a wide plot.
<li style="font-family:Helvica; font-size:21px; "><b>gs[1, :-1]:</b> Occupies the first two columns of the second row.
<li style="font-family:Helvica; font-size:21px; ">Using custom spans enables you to build intricate layouts, making GridSpec highly versatile for complex figures.</br></br>
<p style="font-family:Helvica; font-size:24px; ">

<li style="font-family:Helvica; font-size:21px; "><b>plt.subplots()</b> for more control over each subplot.

<b>a. Adding Legends</b>
<p style="font-family:Helvica; font-size:21px; ">
Legends help differentiate multiple plot elements. Use label within each plot command, then call <b>plt.legend().</b>

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 5, 10, 17, 26]

# Plotting two lines with labels for legend
plt.plot(x, y1, label="Dataset 1", color="blue")
plt.plot(x, y2, label="Dataset 2", color="orange")

plt.title("Line Plot with Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display legend in the upper left corner
plt.legend(loc="upper left")
plt.show()
```

2.<b> Customizing Line Styles, Markers, and Colors</b>
<p style="font-family:Helvica; font-size:21px; ">

```python
# Customized line and marker styles
plt.plot(x, y1, linestyle="--", marker="o", color="purple", label="Dashed Line with Circles")
plt.plot(x, y2, linestyle="-.", marker="s", color="green", label="Dotted Line with Squares")

plt.title("Customized Line Styles and Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="best")
plt.show()
```

<p style="font-family:Helvica; font-size:24px; ">
<b>a. Simple Text Annotation</b>

```python
# Plotting data
plt.plot(x, y1, marker="o", color="blue")
plt.title("Plot with Text Annotations")

# Simple text annotation
plt.text(3, 20, "High Point", color="red", fontsize=12)  # (x, y) position for text

plt.show()
```

<p style="font-family:Helvica; font-size:24px; ">
<b>b. Annotating with Arrows using plt.annotate()</b>
<p style="font-family:Helvica; font-size:21px; ">
Using <b>plt.annotate()</b> allows you to draw an arrow from the text to a specific point for clearer reference.

```python
# Data
y3 = [3, 8, 15, 7, 12]

# Plotting data
plt.plot(x, y3, marker="o", color="purple")
plt.title("Plot with Arrow Annotation")

# Annotation with arrow
plt.annotate("Local Max", xy=(3, 15), xytext=(2, 18),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6),
             fontsize=12, color="darkblue")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>xy=(3, 15):</b> Position of the point to annotate.
<li style="font-family:Helvica; font-size:21px; "><b>xytext=(2, 18):</b> Location where the text will appear.
<li style="font-family:Helvica; font-size:21px; "><b>arrowprops:</b> Defines the appearance of the arrow.
<ul>
<li style="font-family:Helvica; font-size:21px; "><b>facecolor:</b> Color of the arrow.
<li style="font-family:Helvica; font-size:21px; "><b>shrink:</b> Adjusts the arrow length.
<li style="font-family:Helvica; font-size:21px; "><b>width, headwidth:</b> Customize the arrow’s thickness and head size.
</ul>
<p style="font-family:Helvica; font-size:24px; ">
5.<b> Advanced Styling for Annotations with Custom Arrows and Fonts</b>
<p style="font-family:Helvica; font-size:21px; ">
You can further customize <b>plt.annotate()</b> to style annotations with different arrow types and text properties.

```python
# Plotting data with customized annotation
plt.plot(x, y1, color="teal", marker="D", label="Dataset")

# Advanced annotation
plt.annotate("Interesting Point",
             xy=(4, 16), xytext=(3, 22),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3", color="red", lw=2),
             fontsize=12, color="red", fontweight="bold")

# Adding grid and legend
plt.grid(True, linestyle=":", color="gray")
plt.legend()
plt.title("Advanced Annotation with Custom Arrow")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<li style="font-family:Helvica; font-size:21px; "><b>connectionstyle="arc3,rad=.3":</b> Adds a curved arc to the arrow’s path for better visual separation.
<li style="font-family:Helvica; font-size:21px; "><b>fontweight="bold":</b> Makes the annotation text bold.</br></br>

Pair plots show the pairwise relationships between all numerical features in a dataset and are often used in exploratory data analysis (EDA).



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
data = sns.load_dataset("penguins")

# Create a pair plot with hue for species
sns.pairplot(data, hue="species", palette="coolwarm")
plt.suptitle("Pairwise Relationships in Penguin Dataset", y=1.02)
plt.show()
```

<b>b. Correlation Heatmap with sns.heatmap()</b></br>
Heatmaps are useful for visualizing correlation matrices, making it easy to identify strong or weak relationships between variables.

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=["float64", "int64"])

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Plot the correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>select_dtypes(include=["float64", "int64"]):</b></br> This ensures that only columns with numeric data types (integer or float) are selected. If your dataset contains mixed data types, this will filter out non-numeric columns like strings or categorical variables.

<li style="font-family:Helvica; font-size:21px; "><b>correlation_matrix = numerical_data.corr():</b></br> Computes the correlation matrix based on the selected numerical columns only.

<li style="font-family:Helvica; font-size:21px; "><b>sns.heatmap():</b> This will generate the heatmap based on the correlation matrix, where:

<ul><li style="font-family:Helvica; font-size:21px; "><b>annot=True:</b> Adds the correlation values to the heatmap.
<li style="font-family:Helvica; font-size:21px; "><b>cmap="YlGnBu":</b> Uses a yellow-green-blue color map for the heatmap.
<li style="font-family:Helvica; font-size:21px; "><b>vmin=-1, vmax=1:</b> Sets the scale of correlation values to range from -1 to 1 (the possible range for correlation).</ul></br></br>

```python
# Violin plot for body mass across species
sns.violinplot(data=data, x="species", y="body_mass_g",hue="sex", inner="quartile", palette="Set3")
plt.title("Body Mass Distribution by Species")
plt.show()
```

Joint plots combine scatter plots and histograms (or KDE) to show relationships between two variables, along with their marginal distributions.

3.<b> Using FacetGrids for Conditional Subplots</b>
<p style="font-family:Helvica; font-size:21px; ">
FacetGrids allow you to create conditional subplots based on a categorical variable, letting you compare data across different categories in separate plots.</br></br>
<b>Example with sns.FacetGrid()</b></br>
Suppose we want to create a grid of scatter plots for each penguin species to examine how flipper length and body mass vary by sex.

```python
# FacetGrid with scatter plots by species and sex
g = sns.FacetGrid(data, col="species", hue="sex", height=4, aspect=1)
g.map(sns.scatterplot, "flipper_length_mm", "body_mass_g").add_legend()
g.set_axis_labels("Flipper Length (mm)", "Body Mass (g)")
g.set_titles("{col_name}")
plt.suptitle("Flipper Length vs. Body Mass by Species and Sex", y=1.02)
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<li style="font-family:Helvica; font-size:21px; "><b>col="species":</b> Creates one plot per species.

<p style="font-family:Helvica; font-size:24px; ">
1. <b>Time Series Plots in Seaborn</b>
<p style="font-family:Helvica; font-size:21px; ">
Time series plots display data points over time, highlighting trends, cycles, or seasonality. With Seaborn, you can leverage both line plots and additional tools for handling time-based data effectively.</br></br>
<b>Example:</b> Using sns.lineplot() for Time Series</br></br>
The lineplot function in Seaborn is ideal for visualizing trends over time. This plot automatically handles dates as the x-axis if a Pandas datetime object is used.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample time series data
# Generating a sample dataset with dates
dates = pd.date_range(start="2023-01-01", periods=100, freq="D")
data = pd.DataFrame({
    "date": dates,
    "value": pd.Series(100 + (pd.Series(range(100)) * 0.5)).cumsum() + pd.Series(range(100)).apply(lambda x: x % 10)
})

# Plotting the time series
sns.lineplot(data=data, x="date", y="value", color="blue")
plt.title("Time Series Plot of Values Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>

<li style="font-family:Helvica; font-size:21px; "><b>date:</b> A Pandas DatetimeIndex that Seaborn interprets as time-based data.
<li style="font-family:Helvica; font-size:21px; "><b>sns.lineplot():</b> Creates a line chart where x-axis shows dates and y-axis shows the data’s value trend over time.</br></br>
<b>Customizing Time Series Plots</b></br>
You can enhance time series plots by adding markers, adjusting line styles, or changing colors.

```python
# Enhanced time series plot with markers and style adjustments
sns.lineplot(data=data, x="date", y="value", marker="o", linestyle="-", color="darkorange")
plt.title("Enhanced Time Series Plot with Markers")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)  # Rotate dates for better readability
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Additional Options:</b>

<li style="font-family:Helvica; font-size:21px; "><b>marker:</b> Adds markers for each data point (e.g., "o" for circular markers).
<li style="font-family:Helvica; font-size:21px; "><b>linestyle:</b> Changes line style, such as dashed ("--") or solid ("-").
<li style="font-family:Helvica; font-size:21px; "><b>color:</b> Sets line color.</br></br>
<p style="font-family:Helvica; font-size:24px; ">
<b>2. Regression Plots in Seaborn</b>
<p style="font-family:Helvica; font-size:21px; ">
Regression plots visualize the relationship between two variables and fit a regression line, helping assess whether they are correlated. Seaborn provides several tools for this:

<li style="font-family:Helvica; font-size:21px; "><b>sns.regplot():</b> Basic scatter plot with a regression line.
<li style="font-family:Helvica; font-size:21px; "><b>sns.lmplot():</b> More versatile, allowing you to add categories and split into facets.</br></br>
<p style="font-family:Helvica; font-size:21px; ">
<b>a. Simple Regression with sns.regplot()</b></br>
The regplot function creates scatter plots with a regression line, helping show the linear relationship between two variables.

```python
# Load a sample dataset for regression
data = sns.load_dataset("tips")

# Basic regression plot
sns.regplot(data=data, x="total_bill", y="tip", color="teal", line_kws={"color": "orange"})
plt.title("Regression Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>

<li style="font-family:Helvica; font-size:21px; "><b>line_kws={"color": "orange"}:</b> Specifies the color for the regression line.
<li style="font-family:Helvica; font-size:21px; ">The plot shows how tips relate to total bills, providing a regression line to estimate the relationship.</br></br>
<p style="font-family:Helvica; font-size:21px; ">
<b>b. Conditional Regression with sns.lmplot()</b></br>
The <b>lmplot</b> function offers more flexibility for adding categories or splitting data into multiple plots, allowing you to analyze relationships by category.

```python
# Conditional regression by day
sns.lmplot(data=data, x="total_bill", y="tip", hue="day", palette="coolwarm", markers=["o", "x", "s", "D"])
plt.title("Regression of Total Bill vs Tip by Day")
plt.show()
```

<p style="font-family:Helvica; font-size:21px; ">
<b>Explanation:</b>
<li style="font-family:Helvica; font-size:21px; "><b>hue="day":</b> Splits data by day, creating separate regression lines for each day.
<li style="font-family:Helvica; font-size:21px; "><b>palette="coolwarm":</b> Sets a color gradient for each category.
<li style="font-family:Helvica; font-size:21px; "><b>markers:</b> Specifies marker shapes for each group.</br></br>
<p style="font-family:Helvica; font-size:24px; ">
<b>Customizing Regression Plots with Seaborn</b></br>
Seaborn offers additional control over regression settings, such as:
<li style="font-family:Helvica; font-size:21px; "><b>Adding Confidence Intervals:</b> <b>ci=None</b> disables confidence intervals if you want a cleaner look.
<li style="font-family:Helvica; font-size:21px; "><b>Higher-Order Fits:</b> Use <b>order=n</b> for polynomial regression (e.g., order=2 for quadratic).
<li style="font-family:Helvica; font-size:21px; "><b>Logarithmic Regression:</b> Set <b>logx=True</b> to transform the x-axis to a log scale.</br>
<b>Example:</b> Adding Higher-Order Regression

```python
# Quadratic regression (order=2)
sns.regplot(data=data, x="total_bill", y="tip", color="purple", order=2)
plt.title("Quadratic Regression Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
